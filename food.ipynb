```python
def cleaner(x):
    if x=='nan':
        return x.replace('nan','0')
    elif x=='Personal ':
        return x.replace("Personal ",'0')
    elif x=='3.79 bitch':
        return x.replace("3.79 bitch",'0')
    elif x=='Unknown':
        return x.replace("Unknown",'0')
    else:
        return x
```

```python
df['GPA'] = df['GPA'].astype(str)
```

```python
df['GPA'] = df['GPA'].apply(cleaner)
```

```python
df['GPA'].unique()
```

```python
df['GPA'].isnull().sum()
```

```python
df['GPA'] = df['GPA'].astype(float)
```

```python
df['weight'] = df['weight'].replace(["I'm not answering this. "],np.nan)
    
        
```

```python
def clean(x):
    if x=='Not sure, 240':
        return x.split(' ')[2]
    elif x=='144 lbs':
        return x.split(' ')[0]
    else:
        return x
```

```python
df['weight'] = df['weight'].apply(clean)
```

```python
df['weight'].unique()
```

```python
df['weight'] = df['weight'].astype(float)
```

```python
df['weight'] = df['weight'].fillna(df['weight'].median())
```

```python
# Now the missing values  treatment is done
```

```python
categorical_missing = [i for i in df_missing_values if df[i].dtype=='object']
```

```python
numerical_missing = [i for i in df_missing_values if df[i].dtype=='int'or df[i].dtype=='float']
```

```python
def treatnum(x):
    for i in x:
        df[i] = df[i].fillna(df[i].median())
        
```

```python

treatnum(numerical_missing)
```

```python
df[numerical_missing]
```

```python
df[categorical_missing]
```

```python
def treatcat(x):
    for i in x:
        df[i] = df[i].fillna("unknown")
```
